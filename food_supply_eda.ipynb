```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

#### Importing the dataset

```python
#Africa Food Supply
afs_data = pd.read_excel('./Africa Food Supply (2004 - 2013).xlsx', index_col = None, header = 0)
afs_data = afs_data[2:]
afs_data.columns = ["Country","Year","Value"]
```

```python
afs_data
```

#### Check and Remove NaN values

```python
afs_data.isnull().sum()
```

```python
afs_data.dropna(inplace = True)
```

```python
afs_data['Year'] = pd.to_datetime(afs_data['Year'],format='%Y')
afs_data['Year'] = pd.DatetimeIndex(afs_data['Year']).year
afs_data["Value"] = pd.to_numeric(afs_data["Value"])
```

```python
afs_data.info()
```

```python
format_dict = {'Value': '{0:,.0f} kt'}
total_group = afs_data.groupby(["Country","Year"])[["Value"]].sum()
#total_sup = pd.DataFrame(total_group)
total_group.style.format(format_dict)
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=total_group, x='Year', y='Value', hue='Country', style= 'Country')
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x=afs_data["Year"], y = afs_data["Value"])
```

```python
afs_data.groupby(["Year"])[["Value"]].sum().plot.bar(color = 'seagreen')
plt.title("Food Supply in AFrica 2004 - 2013")
plt.show()
```

```python
afs_data.groupby(["Country"])[["Value"]].sum().sort_values(by = 'Value', ascending = False)
```

```python
# filtering = df['Team'] == "Boston Celtics"
# In [15]
#total_group.where(total_group["Year"]==2012)
high = afs_data['Value'].max()
year_sort = afs_data.query('Year==2012').sort_values('Value', ascending = False)
year_sort.style.format({'Value': '{0:,.0f} kt'})
```
